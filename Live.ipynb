```python
a = df_calendar['price'].isna().sum()
b = df_calendar.shape[0]
print(a,b,a/b)
# you make decision here whether to drop or not. 
```

```python
df_calendar['price'] = df_calendar['price'].str.replace('$', '')
df_calendar['price'] = df_calendar['price'].replace(',', '', regex=True)
df_calendar['price']
```

```python
# if decide to drop
df_calendar_dropped = df_calendar.dropna(subset=['price'])
df_calendar_dropped
```

```python
df_calendar_dropped.isna().sum()
```

```python
# cast price to float
df_calendar['price'] = df_calendar['price'].astype(float)
df_calendar_dropped['price'].astype(float)
df_calendar['price']
```

```python
df_calendar_dropped.dtypes
```

```python
df_calendar.isna().sum()
```

```python
# if don't want to drop, replace with mean
# make NaN to 0 first
df_calendar['price'] = df_calendar['price'].fillna(0)
```

```python
df_calendar['price'].isna().sum()
```

```python
df_calendar.dtypes
```

```python
# can fill 0 with mean
mean = df_calendar['price'].mean()
df_calendar['price'] = df_calendar['price'].replace(0, mean)
df_calendar['price']
# can talk about winsorization here, outliers
```
